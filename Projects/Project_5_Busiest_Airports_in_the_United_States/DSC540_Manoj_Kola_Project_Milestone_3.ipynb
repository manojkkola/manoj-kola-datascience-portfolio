### Read HTML tables from the Wikipedia page & Cleaning/Formatting Website Data


This code demonstrates how to extract tabular data from a Wikipedia page using Python's `requests` and `pandas` libraries.

##### Code Overview

The code below scrapes the Wikipedia page listing the busiest airports in the United States and extracts all HTML tables present on that page.

##### Step-by-Step Breakdown

1. **Define the target URL**: We specify the Wikipedia page containing airport traffic data
2. **Set request headers**: Browser headers are added to prevent the request from being blocked
3. **Fetch the webpage**: A GET request retrieves the HTML content
4. **Parse HTML content**: The response is converted into a format compatible with pandas
5. **Extract tables**: `pd.read_html()` automatically identifies and extracts all tables from the page
6. **Verify extraction**: The number of tables found is printed for confirmation

##### Key Components

- **URL**: Wikipedia page for busiest U.S. airports
- **User-Agent header**: Mimics a browser request to avoid potential blocking
- **`pd.read_html()`**: Pandas function that automatically parses HTML tables into DataFrames
- **Output**: A list of DataFrames, each representing a table from the webpage

##### Result

The script outputs the total number of tables extracted, which can then be individually examined and analyzed.

```python
# Define the Wikipedia URL containing the airport data
url = "https://en.wikipedia.org/wiki/List_of_the_busiest_airports_in_the_United_States"

# Create headers to mimic a browser request (helps avoid request blocking)
headers = {"User-Agent": "Mozilla/5.0"}

# Send a GET request to the website and store the response
response = requests.get(url, headers=headers)

# Convert the HTML response text into a readable format for pandas
html = StringIO(response.text)

# Read all tables from the HTML page into a list of DataFrames
tables = pd.read_html(html)

# Count the total number of tables extracted from the Wikipedia page
num_tables = len(tables)

# Print the total number of tables found
print("Total tables extracted from the Wikipedia page:", num_tables)
```

#### Format data into a more readable format
Fix casing and strip extra spaces in textual columns

##### Data Cleaning: Standardize Text Formatting
This code block cleans and standardizes the text formatting across three airport-related datasets (`df_table0`, `df_table1`, `df_table2`).

**Processing Steps:**
- **Airport names**: Converted to title case (e.g., "john f kennedy" → "John F Kennedy")
- **City names**: Converted to title case (e.g., "new york" → "New York")
- **State codes**: Converted to uppercase (e.g., "ny" → "NY")
- **Whitespace**: Extra leading/trailing spaces are removed from all the 3 fields

The code safely checks for column existence before applying transformations, ensuring it works even if some datasets have different structures.

```python
# Loop through three datasets: df_table0, df_table1, df_table2
for df in [df_table0, df_table1, df_table2]:
    
    # If the 'Airport' column exists, convert to a string, remove extra spaces, and title-case it
    if 'Airport' in df.columns:
        df['Airport'] = df['Airport'].astype(str).str.strip().str.title()
    
    # If the 'City' column exists, convert to string, remove extra spaces, and title-case it
    if 'City' in df.columns:
        df['City'] = df['City'].astype(str).str.strip().str.title()
    
    # If the 'State' column exists, convert to string, remove extra spaces, and convert to uppercase
    if 'State' in df.columns:
        df['State'] = df['State'].astype(str).str.strip().str.upper()

# Print confirmation that the step is completed
print("Casing fixed and extra spaces removed for all three datasets.")
```

#### Replace Special characters like '/' & '–' from the Airport field

##### Airport Column Cleaning
This code cleans and standardizes the 'Airport' column across multiple dataframes by removing special characters and applying consistent formatting.


##### What it does
- Iterates through three dataframes: `df_table0`, `df_table1`, and `df_table2`
- Verifies the 'Airport' column exists before processing
- Applies a series of string transformations:
  - Converts all values to strings
  - Replaces forward slashes (/) with spaces
  - Replaces en dashes (–) with spaces
  - Removes extra whitespace
  - Capitalizes the first letter of each word

##### Example
**Before:** `"jfk/new–york  "`  
**After:** `"Jfk New York"`

```python
# Loop through multiple dataframes to clean the 'Airport' column
for df in [df_table0, df_table1, df_table2]:
    
    # Check if the 'Airport' column exists in the current dataframe
    if 'Airport' in df.columns:
        
        # Clean and format the 'Airport' column
        df['Airport'] = (
            df['Airport']
            .astype(str)                  # Convert all values to string
            .str.replace('/', ' ', regex=False)  # Replace slashes with spaces
            .str.replace('–', ' ', regex=False)  # Replace dashes with spaces
            .str.strip()                  # Remove leading/trailing whitespace
            .str.title()                  # Capitalize each word
        )
```

#### Calculate a New Column - Percentage Change Across Multiple DataFrames

This code loops through three dataframes (`df_table0`, `df_table1`, `df_table2`) and calculates the year-over-year percentage change in passenger counts between 2024 and 2023.

**Key Steps:**
- Iterates through each dataframe in the list
- Checks for the presence of both `Passengers_2024` and `Passengers_2023` columns
- Calculates percentage change using the formula: `((2024 - 2023) / 2023) × 100`
- Adds the result as a new column `Pct_Change_24_23` to each dataframe

**Output:** A confirmation message is printed once the percentage change column has been successfully added to all applicable tables.

```python
# Loop through a list of dataframes to calculate percentage change
for df in [df_table0, df_table1, df_table2]:
    
    # Check if both 'Passengers_2024' and 'Passengers_2023' columns exist
    if 'Passengers_2024' in df.columns and 'Passengers_2023' in df.columns:
        
        # Calculate the percentage change between 2024 and 2023 passengers
        df['Pct_Change_24_23'] = (
            (df['Passengers_2024'] - df['Passengers_2023']) 
            / df['Passengers_2023']
        ) * 100  # Multiply by 100 to express as a percentage

# Print confirmation message
print("Added percentage change column (2024 vs 2023) to all applicable tables.")
```

#### Formatting Percentage Change Column

This code loops through multiple dataframes (`df_table0`, `df_table1`, `df_table2`) to format the `Pct_Change_24_23` column. For each dataframe:

1. **Checks** if the `Pct_Change_24_23` column exists
2. **Rounds** the values to 2 decimal places
3. **Converts** to string format and appends a '%' symbol

The formatting ensures consistent percentage display across all three tables, making year-over-year changes (2024 vs 2023) easier to read in the output.

```python
# Loop through multiple dataframes to format the 'Pct_Change_24_23' column
for df in [df_table0, df_table1, df_table2]:
    
    # Check if the column exists in the dataframe
    if 'Pct_Change_24_23' in df.columns:
        
        # Round the values to 2 decimal places, convert to a string, and add a '%' sign
        df['Pct_Change_24_23'] = df['Pct_Change_24_23'].round(2).astype(str) + '%'

# Print confirmation that the formatting was applied
print("Added Pct_Change_24_23 with rounding and % sign.")
```

#### City and State Column Separation

This code processes the `City_Original` column in `df_table2`, which contains combined city and state information, and separates them into distinct columns.

**Steps:**

1. **Split the combined column**: The `City_Original` column is split at the first comma into city and state components
2. **Clean the data**: Both city and state values are stripped of extra whitespace and converted to title case
3. **Insert new columns**: `City` and `State` columns are added immediately after `City_Original`
4. **Standardize territories**: Special handling for Guam and Puerto Rico to ensure consistent naming
5. **Convert state names to codes**: Full state names are mapped to their 2-letter abbreviations using a comprehensive state mapping dictionary

**Result**: The dataframe now contains separate, standardized `City` and `State` columns with proper formatting and state abbreviations.

```python
# Convert 'City_Original' column to string and split at the first comma into two new columns
split_cols = df_table2['City_Original'].astype(str).str.split(',', n=1, expand=True)

# Ensure the resulting DataFrame has exactly two columns labeled 0 and 1
split_cols = split_cols.reindex(columns=[0, 1])

# Display the first few rows to check the result
# split_cols.head()

# Clean the 'city' column - Convert to string, remove extra spaces, and capitalize each word
city_clean = split_cols[0].astype(str).str.strip().str.title()

# Clean the 'state' column - Convert to string, remove extra spaces, and capitalize each word
state_clean = split_cols[1].astype(str).str.strip().str.title()

# Find the column index of 'City_Original' in the dataframe
col_index = df_table2.columns.get_loc('City_Original')

# Insert a new column 'City' right after 'City_Original' with the cleaned city values
df_table2.insert(col_index + 1, 'City', city_clean)

# Insert a new column 'State' after the new 'City' column with the cleaned state values
df_table2.insert(col_index + 2, 'State', state_clean)

# Standardize 'State' values for Guam - If the 'State' column has any variations of 'Guam', set it explicitly to 'Guam'
df_table2.loc[df_table2['State'] == 'Guam', 'State'] = 'Guam'

# Standardize 'State' values for Puerto Rico - If the 'State' column has any variations of 'Puerto Rico', set it explicitly to 'Puerto Rico'
df_table2.loc[df_table2['State'] == 'Puerto Rico', 'State'] = 'Puerto Rico'

# Create a mapping dictionary for U.S. state names to their 2-letter codes
state_map = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR',
    'California': 'CA', 'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE',
    'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID',
    'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS',
    'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
    'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV',
    'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY',
    'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK',
    'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI',
    'South Carolina': 'SC', 'South Dakota': 'SD', 'Tennessee': 'TN',
    'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA',
    'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
}

# Convert full state names in the 'State' column to 2-letter codes
# If a state name is not in the mapping, keep it as is
df_table2['State'] = df_table2['State'].map(state_map).fillna(df_table2['State'])
```
