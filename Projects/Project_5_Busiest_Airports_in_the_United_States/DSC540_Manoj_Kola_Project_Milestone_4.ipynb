#### Step 1: API Calls for All Cities

This section demonstrates how to collect current weather data for multiple locations using the OpenWeatherMap API.

- **Records List**: An empty list is created to store weather data retrieved from the API for each location.
- **Loop Through Locations**: We iterate over a list of locations, extracting the city, state, and country for each entry.
- **Build API URL**: A URL is constructed for each location, specifying the city, state, country, API key, and units (Fahrenheit in this case).
- **API Request**: The `requests` library sends a request to the OpenWeatherMap API and retrieves the response.
- **Convert to JSON**: The API response is converted into JSON format for easier handling.
- **Add Input Metadata**: Original location information (city, state, country) is added to each JSON record to keep track of the source.
- **Append to Records**: Each JSON record is added to the `records` list.
- **Create DataFrame**: The list of JSON records is normalized and converted into a Pandas DataFrame for easy data manipulation and analysis.
- **Display Data**: The DataFrame is displayed to inspect the collected weather data.

```python
# Create an empty list to store API results
records = []

# Loop through each location in the list
for loc in locations:
    
    # Get city, state, and country from dictionary
    city = loc["city"]
    state = loc["state"]
    country = loc["country"]

    # Build the API URL for current weather (units in Fahrenheit)
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city},{state},{country}&appid={API_KEY}&units=imperial"
    
    # Send request to the API
    response = requests.get(url)
    
    # Convert response to JSON format
    data = response.json()

    # Add original input values to the response
    data["input_city"] = city
    data["input_state"] = state
    data["input_country"] = country

    # Add the data to the records list
    records.append(data)

# Convert list of JSON records into a DataFrame
df = pd.json_normalize(records)

# Display the DataFrame
df
```

#### Step 2: Expand Weather Column (Fix Missing Data)

In this step, we transform the nested weather information in our dataset into a more usable format.

- **Extract Weather Data**: The `weather` column contains a list of weather objects. We extract the first object from each list and expand it into separate columns.
- **Rename Columns**: To make the column names more readable, we rename them as:
  - `id` → `Weather_Id`
  - `main` → `Weather_Main`
  - `description` → `Weather_Description`
  - `icon` → `Weather_Icon`
- **Update DataFrame**: We remove the original `weather` column and join the newly expanded columns to the main dataframe.
- **Display Data**: Finally, we display the updated dataframe to verify the changes.

```python
# Extract the first weather object from the list and convert it into separate columns
weather_expanded = df["weather"].apply(lambda x: x[0]).apply(pd.Series)

# Rename the weather columns to make them more readable
weather_expanded = weather_expanded.rename(columns={
    "id": "Weather_Id",
    "main": "Weather_Main",
    "description": "Weather_Description",
    "icon": "Weather_Icon"
})

# Remove the original "weather" column and join the new weather columns to the primary dataframe
df = pd.concat([df.drop(columns=["weather"]), weather_expanded], axis=1)

# Display updated dataframe
df
```

In this step, we improve the readability and consistency of our dataset's column names.

- **Rename Columns**: We assign more meaningful names to each column, e.g., `input_city` becomes `City` and `main.temp` becomes `Temperature_F`.
- **Format Column Names**: We standardize column names by replacing underscores with spaces, capitalizing each word, and then converting spaces back to underscores. This ensures consistency across all columns.

#### Step 4: Standardizing Text Fields in Weather Data

In this section, we clean and standardize the text fields in our weather dataset to ensure consistency and readability:

- **City**: Format the city names so that the first letter of each word is capitalized and any extra spaces are removed.
- **State**: Convert state codes to uppercase and remove extra spaces for uniformity.
- **Country**: Convert country codes to uppercase and remove extra spaces for consistency.
- **Weather Description**: Make weather descriptions more readable by applying title case formatting.
- **Confirmation Message**: A message is printed to confirm that the text fields have been standardized.
- **Display Cleaned Columns**: Show the cleaned `City`, `State`, `Country`, and `Weather_Description` columns for verification.

```python
# Fix city name formatting (make first letter capital and remove extra spaces)
df["City"] = df["City"].str.title().str.strip()

# Convert state to uppercase and remove extra spaces
df["State"] = df["State"].str.upper().str.strip()

# Convert country to uppercase and remove extra spaces
df["Country"] = df["Country"].str.upper().str.strip()

# Make weather description more readable (Title Case)
df["Weather_Description"] = df["Weather_Description"].str.title()

# Print confirmation message
print("Standardized the text fields\n")

# Display cleaned columns
df[["City", "State", "Country", "Weather_Description"]]
```

#### Step 5: Converting Unix Timestamps to Readable Date & Time

In this section, we convert Unix timestamps from the weather dataset into human-readable date and time formats.  

- **Convert Timestamp**: The `Dt` column contains the main data timestamp in Unix format. We convert it to a standard date-time format.  
- **Convert Sunrise & Sunset**: The `Sys.Sunrise` and `Sys.Sunset` columns contain sunrise and sunset times in Unix format. Converting these makes it easier to read and analyze.  
- **Confirmation Message**: After conversion, a message is printed to confirm that the timestamps have been successfully converted.  
- **Display Data**: Finally, we display selected columns, including city, state, country, timestamp, sunrise, and sunset, for verification.

```python
# Convert Unix timestamp to readable date and time
df["Timestamp"] = pd.to_datetime(df["Dt"], unit="s")

# Convert sunrise time to a readable format
df["Sunrise"] = pd.to_datetime(df["Sys.Sunrise"], unit="s")

# Convert sunset time to a readable format
df["Sunset"] = pd.to_datetime(df["Sys.Sunset"], unit="s")

# Print confirmation message
print("Converted the timestamps\n")

# Display selected columns
df[["City", "State", "Country", "Timestamp", "Sunrise", "Sunset"]]
```

#### Step 6: Data Cleaning for Weather Dataset

In this section, we clean the weather dataset to ensure the values are valid and meaningful.

- **Handling Humidity Values**: Any humidity percentage values that are less than 0% or greater than 100% are invalid. These values are replaced with `NaN` to indicate missing values.  
- **Handling Wind Speed Values**: Wind speed cannot be negative, so any negative values are replaced with `NaN`.  
- **Confirmation Message**: A simple message is printed to indicate that the invalid values have been cleaned successfully.  
- **Verification**: Selected columns, including `City`, `State`, `Country`, `Humidity_Pct`, and `Wind_Speed_Mph`, are displayed to visually verify the cleaning process.

```python
import numpy as np  # Make sure numpy is imported

# Replace invalid humidity values (<0 or >100) with NaN
df["Humidity_Pct"] = df["Humidity_Pct"].apply(
    lambda x: np.nan if (x < 0 or x > 100) else x
)

# Replace invalid wind speed values (<0) with NaN
df["Wind_Speed_Mph"] = df["Wind_Speed_Mph"].apply(
    lambda x: np.nan if x < 0 else x
)

# Print a message to show that data cleaning is done
print("Cleaned the invalid values\n")

# Step 4: Display selected columns to verify cleaning
df[["City", "State", "Country", "Humidity_Pct", "Wind_Speed_Mph"]]
```

#### Step 7: Deriving Additional Weather Metrics

In this section, we create new columns in our weather dataset to provide more insights:

- **Temperature Range (Fahrenheit)**: Calculates the difference between the maximum and minimum temperatures for each entry. This helps identify the daily temperature variation.

- **Daytime Indicator**: Determines whether a given timestamp falls during daytime. It checks if the timestamp is between the recorded sunrise and sunset times, returning `True` if it is daytime and `False` otherwise.

- **Confirmation Message**: Prints a message to indicate that the derived columns have been successfully added to the dataset.

- **Displaying Key Columns**: Shows important columns including `City`, `State`, `Country`, and the newly created `Temp_Range_F` and `Is_Daytime` for review.

```python
# Calculate temperature range in Fahrenheit (max - min)
df["Temp_Range_F"] = df["Temp_Max_F"] - df["Temp_Min_F"]

# Create a column to check if the timestamp is during the daytime
# True if current time is between sunrise and sunset, else False
df["Is_Daytime"] = df.apply(
    lambda row: row["Sunrise"] <= row["Timestamp"] <= row["Sunset"],
    axis=1
)

# Print a message to indicate that new columns were added
print("Added the derived columns\n")

# Display relevant columns, including the newly created ones
df[["City", "State", "Country", "Temp_Range_F", "Is_Daytime"]]
```

Transforming nested JSON and deriving new metrics could also introduce errors if not carefully verified.

##### Ethical Mitigation Measures
To address potential ethical concerns, all transformation steps were clearly documented, and the original raw data fields were preserved before cleaning. Each assumption and its justification were transparently stated. Cross-verification of API data and careful handling of outliers were used to minimize the risk of errors. These measures ensure that the final dataset is both usable and ethically responsible, while maintaining the credibility and integrity of the information.
