#### Medium 1: PowerPoint Presentation

**Audience:**  
State policymakers and executive decision-makers (non-technical)

**Plan:**  
This presentation will focus on high-level insights about childcare costs across states and over time. The goal is to clearly communicate affordability challenges and their impact on families using simple visuals and minimal text.

**Storyboard (Slide Plan):**
1. Introduction – Objective of the analysis  
2. Overview of Dataset (Childcare pricing trends)  
3. Childcare Cost Trends Over Time (Line Chart)  
4. State-wise Cost Comparison (Bar Chart)  
5. Key Insights (Rising costs, regional differences)  
6. Impact on Families & Economy  
7. Recommendations & Call to Action (Policy changes, subsidies) 

---

#### Medium 2: Interactive Dashboard (Python / Power BI / Tableau)

**Audience:**  
Policy analysts, researchers, and government agencies (moderate data knowledge)

**Plan:**  
This dashboard will allow users to explore childcare costs interactively across states, years, and care types. It will include drill-down features and filters for deeper analysis. Visualizations will be more detailed and **different from the PowerPoint**.

**Mock-up Layout:**

- **Top Left:** Heatmap (Childcare costs by State & Year)  
- **Top Right:** Scatterplot (Cost vs Household Income)  
- **Bottom Left:** Stacked Bar Chart (Cost by Care Type & Year)  
- **Bottom Right:** Distribution Plot (Variation in childcare costs across states)  
- **Filters:** State, Year, Care Type, Income Level  

---

#### Medium 3: Infographic

**Audience:**  
General public/parents (non-technical)

**Plan:**  
This infographic presents a simplified, visually engaging summary of childcare affordability. It will highlight key statistics and insights using icons and minimal text, avoiding technical complexity and overlapping visuals.

**Mock-up:**

- **Top:** Title – “The Rising Cost of Childcare in the U.S.”  
- **Middle:** Key statistics (Average cost, most expensive states)  
- **Lower Section:** Simple comparisons (High vs Low cost states)  
- **Visual Elements:** Icons representing families, costs, and regions  
- **Bottom:** Call to Action (Need for affordable childcare policies)

#### Narrative

This project analyzes the National Database of Childcare Prices to understand trends in childcare affordability across the United States. The findings are presented using three distinct mediums tailored to different audiences. The PowerPoint presentation is designed for policymakers and focuses on high-level trends such as rising childcare costs and state-level differences using simple visualizations. The interactive dashboard, built with Python, Power BI, or Tableau, is intended for analysts and provides deeper insights through advanced visualizations such as heatmaps, scatterplots, and stacked bar charts, along with filtering capabilities for detailed exploration. The infographic targets the general public by summarizing key findings in a visually engaging, easy-to-understand format, using icons and key statistics.

Each medium uses a unique set of visualizations to avoid overlap and ensure clarity, as emphasized in the feedback. The goal of this project is to highlight the increasing financial burden of childcare and its impact on families. The analysis supports a call to action for policymakers to consider strategies such as subsidies, cost regulation, or expanded childcare support programs to improve affordability and accessibility.
